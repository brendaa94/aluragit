```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier #algoritimo de classificação
from sklearn.model_selection import train_test_split #Função para separar banco de dados em treino e validação
```

```python
dados = pd.read_csv ('/content/train.csv')
```

```python
dados.head()
```

```python
dados = dados.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis = 1)
```

```python
dados = dados.set_index(['PassengerId']) #Criação chave primária correspondente ao passageiro
dados = dados.rename(columns = {'Survived' : 'Target'}, inplace = False) #inplace: atribuir nov nome à coluna ou não
```

```python
dados['Sex_F'] = np.where(dados['Sex'] == 'female',1,0)
dados['P_class_2']= np.where(dados['Pclass'] == 1,1,0)
dados['P_class_2']= np.where(dados['Pclass'] == 2,1,0)
dados['P_class_3']= np.where(dados['Pclass'] == 3,1,0)
```

```python
dados = dados.drop(['Pclass','Sex'], axis=1)
```
